### Exercise 1

How many words are there in `text2` from `nltk.book`? How many distinct words are there? Calculate the lexical diversity.


```python
# Count total words
total_words = len(text2)

# Count distinct words
distinct_words = len(set(text2))

# Calculate lexical diversity
lexical_diversity = distinct_words / total_words

# Print the output
print("Total words:",total_words)
print("Total distinct words:",distinct_words)
print("Lexical diversity:",lexical_diversity)
```

### Exercise 4

Use only the `index()` function to find all the indexes of the word "sunset" in `text9`.

```python
indexes = []
current_index = 0

# Loop until all occurrences are found
while True:
    try:
        # Find the next index of the word
        found_index = text9[current_index:].index('sunset')
        actual_index = current_index + found_index
        indexes.append(actual_index)
        
        # Move to the next word
        current_index = actual_index + 1
        
    except ValueError:
        # No more occurrences
        break

print(indexes)
```

### Exercise 7

Find all the four-letter words in `text6`. With the help of a frequency distribution (`FreqDist`), show these words in decreasing order of frequency.


```python
# First, filter out all four-letter words in text6
four_letter_words = [w for w in text6 if len(w) == 4]

# Now create a frequency distribution of these words
fdist = FreqDist(four_letter_words)

# Get the frequency distribution in decreasing order
sorted_fdist = sorted(fdist.items(), key=lambda x: x[1], reverse=True)

sorted_fdist
```

### Exercise 12

Define a function `freq(word, text)` that calculates how often a given word occurs in a text, not using `count()` but a `FreqDist`. Use the function to calculate how often "promise" appears in `text4`.

```python
# Define the function freq
def freq(word, text):
    fdist = FreqDist(text)
    return fdist[word]

# Calculate how often "promise" appears in text4
frequency_of_promise = freq("promise", text4)

frequency_of_promise
```
